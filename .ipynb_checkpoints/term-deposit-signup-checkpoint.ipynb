```python
#Function to group a feature to certain categories and plot a bar graph of the number of cleints subscribed for term deposit
def BucketedSubscribedBarPlots(dataset, feature, buckets, titleSuffix ='', xLabelSuffix = '', yLabelSuffix = '', dropCutsColumn = True):
    #Create the cuts based on the input buckets
    data_cuts = pd.cut(dataset[feature], buckets)
    #print(data_cuts)
    #Define a feature name for this cuts
    feature_cuts = feature+'_cuts'
    # Add the data cuts to the original data set
    dataset[feature_cuts] =  data_cuts
    print('Counts of clients in the bucket of the Feature ', feature, ':\n',  dataset[feature_cuts].value_counts().sort_index())
    #Get the mean of subscribed for the feature cuts. Since the column y is changed to 0 for no and 1 for yes, the mean will tell how may subcribed. 
    data_cuts_by_bucket_group = (dataset.groupby([feature_cuts]).y.mean().reset_index()).dropna()
    #print(bank_marketing_by_age_group)
    #drop the newely created column
    if dropCutsColumn:
        bank_marketing_intial.drop(columns=[feature_cuts], inplace = True)
    # Plot the mean values for each bucket
    plt.figure(figsize=(10, 5))
    bucketBarPlot = sns.barplot(data_cuts_by_bucket_group, x = feature_cuts, y = 'y')
    bucketBarPlot.set_ylabel('% of clients Subscribed for term deposit')
    bucketBarPlot.set_xlabel(feature.title() + xLabelSuffix)
    bucketBarPlot.tick_params(axis='x', rotation=75)
    bucketBarPlot.set_title("Clients subscribed by " + feature.title()+ titleSuffix + " groups")
    #sns.barplot( x = dataset[feature_cuts], y =  dataset[feature_cuts].value_counts())
#SetAxesLabels(dispAge, 'Age', 'Age', 'Probability')
    plt.show(plt)
```

```python
def PlotCategoricaBar(feature):
    #Get the total count by feature requested
    group_by_y = (bank_marketing_intial.groupby(feature).agg({'y':'size'}).rename(columns={'y':'count'}).reset_index())
    #Get the mean.. since the subscribed is changed to 1, the mean of all values will give the number of subscribed
    group_by_y_mean = (bank_marketing_intial.groupby(feature).agg({'y':'mean'}).rename(columns={'y':'mean'}).reset_index())
    #Combining and creating new column of number subscribed. 
    group_by_y['mean'] = group_by_y_mean['mean']
    group_by_y['subscribed'] = group_by_y['mean'] * group_by_y['count']
    print(group_by_y)
    fig, ax = plt.subplots(figsize=(10, 6))
    axb = group_by_y.plot(x=feature, y=["count", "subscribed"], kind="bar", ax = ax)
    ax.set_title('Total number clients in each category vs count of subscribed')
    ax.set_xlabel(feature.title())
    ax.set_ylabel('Number of Clients')
    #ax.bar_label(ax.containers[0])
    #atwin_alt = ax.get_shared_y_axes().get_siblings(ax)
    #print(atwin_alt)
    #atwin_alt.bar_label(atwin_alt.containers[0])
    #Updating the legend values
    legend = plt.legend()
    for xx in legend.get_texts():
        if xx.get_text() == 'count':
            xx.set_text('Clients')
        else:
            xx.set_text('Clients subscribed')
    plt.show(plt)     
```

```python
#VIF function to see multicollinearity
#Making use of the Sample Code provide by Savio
def sk_vif(exogs, data):
  vif_dict = {}

  for exog in exogs:
    not_exog = [i for i in exogs if i !=exog]
    # split the dataset, one independent variable against all others
    X, y = data[not_exog], data[exog]

    # fit the model and obtain R^2
    r_squared = LinearRegression().fit(X,y).score(X,y)
    if r_squared != 1: 
        # compute the VIF
        vif = 1/(1-r_squared)
        vif_dict[exog] = vif

  return pd.DataFrame({"VIF": vif_dict})
```

### Data Preparation

```python
#Reloading the data and converting the classification column to numeric. 
bank_marketing_intial = pd.read_csv('data/bank-full.csv', sep=';')
#Coverting class variabe values from yes and no to 1 and 0 
bank_marketing_intial = bank_marketing_intial.replace({'y': {'yes': 1, 'no': 0}})
```

```python
#Dropping features that are decided to be not used.
bank_marketing_cleaned = bank_marketing_intial.drop(columns=['month', 'poutcome'])
```

```python
bank_marketing_cleaned.head()
```

```python
bank_marketing_cleaned.info()
```

```python
#Dropping data where the balance is greater than 70000 or negative balance of over 5000 (less than -5000)
bank_marketing_cleaned = bank_marketing_cleaned[
    (bank_marketing_cleaned['balance'] >= -5000) & (bank_marketing_cleaned['balance'] <= 70000) ]
print("Number of data records remaning ", bank_marketing_cleaned.shape[0])
```

```python
#Dropping data where the call time is zero greater than 3500 seconds
bank_marketing_cleaned = bank_marketing_cleaned[(bank_marketing_cleaned['duration'] > 0) & (bank_marketing_cleaned['duration'] <= 3500) ]
print("Number of data records remaning ", bank_marketing_cleaned.shape[0])
```

```python
#Dropping data where number of calls before the current campagin are greater than 40
bank_marketing_cleaned = bank_marketing_cleaned[(bank_marketing_cleaned['campaign'] <=40) ]
print("Number of data records remaning ", bank_marketing_cleaned.shape[0])
```

### Modelling

```python
#Splliting data for train and test
X_train, X_test, y_train, y_test = train_test_split(bank_marketing_cleaned.drop(['y'], axis = 1), 
                                                    bank_marketing_cleaned['y'], test_size = 0.3,
                                                   random_state=42)
```

```python
# Creating a column transformer
column_transformer = make_column_transformer(
    (OneHotEncoder(drop="if_binary"),
     bank_marketing_cleaned.select_dtypes(include='object').columns),
    verbose_feature_names_out=False,
    remainder=StandardScaler())
```

###### Reusable Method for GirdSearchCV for the Classifier and Plot Confusion Matrix and ROC plot

```python
def evaluateModel(model, model_params, modelType, col_transformer, X_train, X_test, y_train, y_test, n_jobs_=4):
    ##Build the pipe line with Transformer and Model. 
    pipe          =  Pipeline([('transform', col_transformer),('model', model)])
    
    #Create the Grid for accuracy scorring and recall scoring
    gridCV        =  GridSearchCV(pipe, param_grid=model_params, n_jobs=4)
    gridCV_recall =  GridSearchCV(pipe, param_grid=model_params, scoring = 'recall', n_jobs=n_jobs_)
    start_time = process_time()
    
    #Fit for better accuracy 
    gridCV.fit(X_train, y_train)
    
    #Fit for better recall
    gridCV_recall.fit(X_train, y_train)
    
    ## Retrive the results
    acc_best_params = list(gridCV.best_params_.values())
    acc_best_est    = gridCV.best_estimator_
    acc_score       = acc_best_est.score(X_test, y_test)
    
    recall_best_params = list(gridCV_recall.best_params_.values())
    recall_best_est    = gridCV_recall.best_estimator_
    recal_score       = recall_best_est.score(X_test, y_test)
    
    # Predict on the test data
    acc_best_est_predict    = acc_best_est.predict(X_test)
    recall_best_est_predict = recall_best_est.predict(X_test)
    
    #Plot the Conussion matrix for the best model with accuraccy scoring, bbest model identified by recall scoring    
    fig, ax = plt.subplots(2, 2, figsize=(10,10), constrained_layout = True)
    #accuracy scoring
    accu_confusion_matrix  = confusion_matrix(y_test, acc_best_est_predict, labels=gridCV.classes_)
    ConfusionMatrixDisplay(confusion_matrix=accu_confusion_matrix ,display_labels = ['No', 'Yes']).plot(ax=ax[0, 0])
    #disp.plot()
    ax[0, 0].set_title("Best Model")
    #Recall scoring
    recall_confusion_matrix  = confusion_matrix(y_test, recall_best_est_predict, labels=gridCV_recall.classes_)
    ConfusionMatrixDisplay(confusion_matrix=recall_confusion_matrix ,display_labels = ['No', 'Yes']).plot(ax=ax[0, 1])
    #disp.plot()
    ax[0, 1].set_title("Best Model - GridSearchCV Recall scoring")
    
    #Build the scoring results to a data frame
    model_data = {'Model_Results': ['GridCV Accuracy Scoring Best Params',
                                    'GridCV Recall scoring Best Params',
                                    'Accuracy of GridCV with Accuracy Scoring',
                                    'Recall score of GridCV with Accuracy Scoring',
                                    'Accuracy of GridCV with Recall Scoring',
                                    'Recall score of GridCV with Recall Scoring'
                                   ],
                  modelType: [acc_best_params, recall_best_params, acc_score,
                              recall_score(y_test, acc_best_est_predict),
                              recal_score,
                              recall_score(y_test, recall_best_est_predict)
                             ]} 
    
    results_df = pd.DataFrame(model_data).set_index('Model_Results')
    
    #ROC curve for both the accuracy and recall best model
    RocCurveDisplay.from_predictions(y_test, acc_best_est_predict, ax = ax[1, 0], label = 'Best Model')
    ax[1, 0].grid()
    ax[1, 0].plot(np.arange(0, 1.1, .1), np.arange(0, 1.1, .1), label = 'baseline');
    ax[1, 0].set_title('RocCurveDisplay')
    ax[1, 0].legend(loc='upper center', bbox_to_anchor=(0.5, 1.20));
    
    RocCurveDisplay.from_predictions(y_test, recall_best_est_predict, ax = ax[1, 1], label = 'Best Model - GridSearchCV Recall scoring')
    ax[1, 1].grid()
    ax[1, 1].plot(np.arange(0, 1.1, .1), np.arange(0, 1.1, .1), label = 'baseline');
    ax[1, 1].set_title('RocCurveDisplay- GridSearchCV Recall scoring')
    ax[1, 1].legend(loc='upper center', bbox_to_anchor=(0.5, 1.20));
    
    plt.suptitle(modelType)
    plt.show()
    return results_df, gridCV, gridCV_recall
```

###### Reusable Method for updating the predicts with a threshold and plot Confusion Matrix 

```python
# Function to predict at a threshold and plot the confusion matrix. 
def plotConfusionMatrixWithThreshold(X_test, y_test, threshold, gridCV, modelType):
    
    # Use the probability predict values and set the predictions bases on threshold
    threshold_preds  = np.where(gridCV.best_estimator_.predict_proba(X_test)[:, 1] >= threshold, 1, 0)

    #Plot the confussion matrix with predictions based on thresholds 
    fig, ax = plt.subplots(figsize=(6,5), constrained_layout = True)
    threshold_confusion_matrix  = confusion_matrix(y_test, threshold_preds, labels=gridCV.classes_)
    ConfusionMatrixDisplay(confusion_matrix=threshold_confusion_matrix ,display_labels = ['No', 'Yes']).plot(ax=ax)
    #disp.plot()
    ax.set_title(modelType+"\nConfusion Matrix with threshold: "+ str(threshold))
    
    # Build the data frame with scores based on the threshold predicted out come 
    model_data = {'Model_Results': ['Threshold used',
                                    'Accuracy Score - Threshold',
                                    'Recall Score - Threshold',
                                    'Precision Score - Threshold'
                                   ],
                  modelType: [ str(threshold), accuracy_score(y_test, threshold_preds),
                              recall_score(y_test, threshold_preds),
                              precision_score(y_test, threshold_preds)
                             ]} 
    
    
    results_df = pd.DataFrame(model_data).set_index('Model_Results')
    plt.show()
    
    return results_df, fig
```
